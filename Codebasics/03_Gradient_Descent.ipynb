```python
import numpy as np
```

```python
def gradient_descent(x,y):
  m_current=b_current=0 # We need to start with some value of m & b to reach the global minima
  iterations = 10000 # We can fine tune it later , We need to see how many steps we are planning to take
  n=len(x) # Length of data point
  learning_rate=0.008 # This value needs to be adjusted
  for i in range(iterations):
    y_predicted=m_current*x+b_current # We need to calculate the predicted value of y i.e. y = mx+b 
    cost=(1/n)*sum([val**2 for val in(y-y_predicted) ])
    md=-(2/n)*sum(x*(y-y_predicted))   # Calculate m derivative
    bd=-(2/n)*sum(y-y_predicted)       # Calculate b derivative
    m_current = m_current-learning_rate*md
    b_current = b_current-learning_rate*bd
    print("m {},b {},cost {} ,iteration {}".format(m_current,b_current,cost,i))

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

```

```python
gradient_descent(x,y)
```

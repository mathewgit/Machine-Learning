```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Data Science/Machine Learning/Code Basics/homeprices_onehotencoding.csv.csv")
df
```

```python
pd.get_dummies(df.town) # 
```

```python
dummies = pd.get_dummies(df.town)
```

```python
merged = pd.concat([df,dummies],axis='columns') # Concatanate 
merged
```

```python
final = merged.drop(['town','Unnamed: 3','west windsor'], axis='columns')
final
#Dummy Variable Trap
#When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

#NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Creating X by droping price, because price is a dependent variable
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
```

```python
model.fit(X,y) # Training
```

```python
model.predict(X) # 2600 sqr ft home in new jersey
```

```python
model.predict([[2800,0,1]]) # Area , 0 and 1 robinson
```

```python
model.predict([[3400,0,0]]) # Area , 0 and 1 robinson
```

```python
model.score(X,y) # Mean 95% accurate
```

```python
#Using sklearn OneHotEncoder # We are doing the same thing as above
#First step is to use label encoder to convert town names into numbers
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() # creating a class object
```

```python
dfle = df # creating a new dataframe
dfle.town = le.fit_transform(dfle.town) # It takes town as labels and convert it to 0,1,2
dfle
```

```python
# Creating a training dataset , X is the training dataset
X = dfle[['town','area']].values # Using values because we need x as a 2d array and not a dataframe
X
```

```python
y = dfle.price.values 
y
```

```python
# Importing one hot encoding

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough') # Creating a object
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:] # Take all the rows , All the columns from index 1 onwards
```

```python
X
```

```python
model.fit(X,y)
```

```python
model.predict([[0,1,3400]]) # 3400 sqr ft home in west windsor
```

```python
model.predict([[1,0,2800]]) # 2800 sqr ft home in robbinsville
```
